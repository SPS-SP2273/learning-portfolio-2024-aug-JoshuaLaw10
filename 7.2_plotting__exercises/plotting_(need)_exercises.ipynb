```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='x', markerfacecolor='blue',
             ecolor='green',capsize=5,
             label='Y against X with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.linspace(0, 50, 1000000)
y = (np.e ** (-x/10)) * np.sin(x)

plt.plot(x, y)
plt.xlabel('x',fontsize = 15,color = 'darkorange')
plt.ylabel('y',fontsize = 15,color = 'darkorange',rotation = 0)
plt.title('$e^{-x/10}\\sin(x)$',fontsize = 20)
plt.grid(visible = True, axis ='both',alpha =0.25)
plt.ylim(-1,1)

plt.show()
```
